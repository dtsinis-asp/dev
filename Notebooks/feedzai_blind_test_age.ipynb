```python
import sys
import pandas as pd
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score
```

```python
blind_test = pd.read_csv("Result_118.csv")  

#63 for original
```

```python
#create cols 


tenure_buckets = ['0-30','31-60','61-90','91-180','181-365','365+']

for bucket in tenure_buckets:
  
    eval_df = blind_test[blind_test['TENURE_BUCKETS']==bucket]
#     eval_df = blind_test[['TRANSACTION_ID','LABEL','AMOUNT','RISK_600']]

    eval_df_tp_amount = eval_df[(eval_df['LABEL']==1)&(eval_df['RISK_900']==1)]['AMOUNT'].sum()
    eval_df_tn_amount = eval_df[(eval_df['LABEL']==0)&(eval_df['RISK_900']==0)]['AMOUNT'].sum()
    eval_df_fp_amount = eval_df[(eval_df['LABEL']==0)&(eval_df['RISK_900']==1)]['AMOUNT'].sum()
    eval_df_fn_amount = eval_df[(eval_df['LABEL']==1)&(eval_df['RISK_900']==0)]['AMOUNT'].sum()

    CM = confusion_matrix(eval_df['LABEL'], eval_df['RISK_900'])
    TN = CM[0][0]
    FN = CM[1][0]
    TP = CM[1][1]
    FP = CM[0][1]


    print("-------------")
    print("bucket set to: " + str(bucket))
    print("size of bucket: " + str(len(eval_df)))
    print("Accuracy -->", accuracy_score(eval_df['LABEL'],eval_df['RISK_900']))
    print("Precison -->", precision_score(eval_df['LABEL'],eval_df['RISK_900']))
    print("Recall -->"  , recall_score(eval_df['LABEL'],eval_df['RISK_900']))
    print("False Positive Rate -->", (FP/(FP+TN)))
    print("Number of True Negatives -->",TN)
    print("Number of True Positives -->",TP)
    print("Number of False Negatives -->",FN)
    print("Number of False Positives -->",FP)
    print("tp cost ", eval_df_tp_amount)
    print("tn cost ", eval_df_tn_amount)
    print("fp cost ", eval_df_fp_amount)
    print("fn cost ", eval_df_fn_amount)
```
